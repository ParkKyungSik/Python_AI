```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

def f2(x0,x1):
    r = np.sqrt(x0**2 + x1**2)
    ans = np.sin(r)
    return ans
```

```python
xn = 50
x0 = np.linspace(-5,5,xn)
x1 = np.linspace(-5,5,xn)

xx0,xx1 = np.meshgrid(x0,x1)

x2 = f2(xx0,xx1)

plt.figure(figsize = (10, 7))

ax = plt.subplot(1, 1, 1, projection = '3d')
ax.plot_surface(xx0, xx1, x2, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
ax.set_zticks((0,1))
ax.view_init(30, 70)

plt.title("Mexican Hat Graph")
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.xlabel('$x_0$')
plt.ylabel('$x_1$')
plt.grid(True)
plt.show()
```
