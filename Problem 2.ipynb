```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f3(x0, x1):
    r = np.sqrt(x0**2 + x1**2)
    x2 = np.sin(r)
    return x2

xn = 70
x0 = np.linspace(-5.0, 5.0, xn)
x1 = np.linspace(-5.0, 5.0, xn)
y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])


xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(figsize=(10, 7))
ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')
ax.set_zticks((0, 1))
ax.view_init(30, 70)
ax.set_title('problem2')       
ax.set_xlabel('$x_0$')                           
ax.set_ylabel('$x_1$')                           
ax.set_zlabel('$x_3$') 
plt.show()
```
