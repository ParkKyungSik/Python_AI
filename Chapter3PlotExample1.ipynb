```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f2(x,w):
    return (x-w)*x*(x+2)

x = np.linspace(-3,3,100) 
y = f2(x,2)
z = f2(x,1)

w = y.shape
print(w)


plt.plot(x, y, color='black', label ='$w=2$') # $안에 있는 문자열을 Text취급(특수문자나 함수)
plt.plot(x, z, color='cornflowerblue', label='$w=1$')

plt.legend(loc="upper left") #location(loc)에 위치 지정

plt.ylim(-15,15)
plt.title('$f_2(x)$') # _는 아래첨자
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()

```
