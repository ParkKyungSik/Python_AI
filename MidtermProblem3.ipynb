```python
import numpy as np
import matplotlib.pyplot as plt

delta = 0.01

def f(x0, x1): # (A) f 
    return x0**2 + 2 * x0 * x1 +  3

def df_dx(x0, x1): # (B) f w0  
    dx0 = 2 * x0 + 2 * x1
    dx1 = 2 * x0 
    return dx0, dx1

def delf_dx(x0, x1): # (B) f w0    
    dx0 = (f(x0+delta,x1) - f(x0,x1))/delta
    dx1 = (f(x0,x1+delta) - f(x0,x1))/delta
    return dx0, dx1


x_range = 2
dx = 0.25
x0 = np.arange(-x_range, x_range + dx, dx)
x1 = np.arange(-x_range, x_range + dx, dx)

xn = x0.shape[0]
xx0, xx1 = np.meshgrid(x0, x1) # (D)


ff = f(xx0,xx1)
dff_dx  = df_dx(xx0, xx1)

    
plt.figure(figsize=(24, 6))
plt.subplots_adjust(wspace=0.2)
plt.subplot(1, 3, 1)
cont = plt.contour(xx0, xx1, ff, 10, colors='k') # (F) f  

cont.clabel(fmt='%2.0f', fontsize=8)
plt.xticks(range(-x_range, x_range + 1, 1))
plt.yticks(range(-x_range, x_range + 1, 1))
plt.xlim(-x_range - 0.5, x_range + .5)
plt.ylim(-x_range - .5, x_range + .5)
plt.title('$f(x_0, x_1)$ Graph')
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)


plt.subplot(1, 3, 2)
plt.quiver(xx0, xx1, dff_dx[0], dff_dx[1]) # (G) f   
plt.title('Gradient $f(x_0, x_1)$ Graph using differentiation')
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.xticks(range(-x_range, x_range + 1, 1))
plt.yticks(range(-x_range, x_range + 1, 1))
plt.xlim(-x_range - 0.5, x_range + .5)
plt.ylim(-x_range - .5, x_range + .5)


delf_dx = delf_dx(xx0, xx1)
plt.subplot(1, 3, 3)
plt.quiver(xx0, xx1, delf_dx[0], delf_dx[1]) # (G) f   
plt.title('Gradient $f(x_0, x_1)$ Graph using derived function')
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.xticks(range(-x_range, x_range + 1, 1))
plt.yticks(range(-x_range, x_range + 1, 1))
plt.xlim(-x_range - 0.5, x_range + .5)
plt.ylim(-x_range - .5, x_range + .5)

plt.show()
```
