```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f2(x,w):
    return (x-w)*x*(x+2) #x가 벡터, w가 스칼라

x = np.linspace(-4,4,100)

plt.figure(figsize=(10,3))
plt.subplots_adjust(wspace=0.5, hspace=0.5) #그래프 사이 간격

#for i in range(6):
plt.subplot(2,3,5)
plt.title(5)
plt.plot(x, f2(x,5), 'k')
plt.ylim(-20,20)
plt.grid(True)

plt.subplot(2,3,1)
plt.title(1)
plt.plot(x, f2(x,1), 'k')
plt.ylim(-20,20)
plt.grid(True)

plt.subplot(2,3,3)
plt.title(3)
plt.plot(x, f2(x,3), 'k')
plt.ylim(-20,20)
plt.grid(True)

plt.subplot(2,3,2)
plt.title(2)
plt.plot(x, f2(x,2), 'k')
plt.ylim(-20,20)
plt.grid(True)

plt.subplot(2,3,4)
plt.title(4)
plt.plot(x, f2(x,4), 'k')
plt.ylim(-20,20)
plt.grid(True)

plt.show()
```
